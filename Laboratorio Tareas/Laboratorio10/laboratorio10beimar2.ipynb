#### 1.2.1 Fucin Sigmoidea

La hipotesis para la regresin logistica se define como:

$$ h_\theta(x) = g(\theta^T x)$$

donde la funcin $g$ is la funcin sigmoidea. La funcin sigmoidea se define como: 

$$g(z) = \frac{1}{1+e^{-z}}$$.

Los resultados que debe generar la funcion sigmoidea para valores positivos amplios de `x`, deben ser cercanos a 1, mientras que para valores negativos grandes, la sigmoide debe generar valores cercanos 0. La evaluacion de `sigmoid(0)` debe dar un resultado exacto de 0.5. Esta funcion tambien debe poder trabajar con vectores y matrices.

```python
def sigmoid(z):
    # Calcula la sigmoide de una entrada z
    # convierte la intrada a un arreglo numpy
    z = np.array(z)
  
    g = np.zeros(z.shape)

    g = 1 / (1 + np.exp(-z))

    return g
```

#### 1.2.2 Funcin de Costo y Gradiente

Se implementa la funcion cost y gradient, para la regresin logistica. Antes de continuar es importante agregar el termino de intercepcion a X. 

```python
# Configurar la matriz adecuadamente, y agregar una columna de unos que corresponde al termino de intercepcin. 
m, n = X.shape
# Agraga el termino de intercepcin a A
X = np.concatenate([np.ones((m, 1)), X], axis=1)
```

La funcion de costo en una regresin logistica es:

$$ J(\theta) = \frac{1}{m} \sum_{i=1}^{m} \left[ -y^{(i)} \log\left(h_\theta\left( x^{(i)} \right) \right) - \left( 1 - y^{(i)}\right) \log \left( 1 - h_\theta\left( x^{(i)} \right) \right) \right]$$

y el gradiente del costo es un vector de la misma longitud como $\theta$ donde el elemento $j^{th}$ (para $j = 0, 1, \cdots , n$) se define como:

$$ \frac{\partial J(\theta)}{\partial \theta_j} = \frac{1}{m} \sum_{i=1}^m \left( h_\theta \left( x^{(i)} \right) - y^{(i)} \right) x_j^{(i)} $$

Si bien este gradiente parece idntico al gradiente de regresin lineal, la frmula es diferente porque la regresin lineal y logstica tienen diferentes definiciones de $h_\theta(x)$.
<a id="costFunction"></a>

```python
def calcularCosto(theta, X, y):
    # Inicializar algunos valores utiles
    m = y.size  # numero de ejemplos de entrenamiento

    J = 0
    h = sigmoid(X.dot(theta.T))
    J = (1 / m) * np.sum(-y.dot(np.log(h)) - (1 - y).dot(np.log(1 - h)))
    
    return J
```

```python
def descensoGradiente(theta, X, y, alpha, num_iters):
    # Inicializa algunos valores
    m = y.shape[0] # numero de ejemplos de entrenamiento
    
    # realiza una copia de theta, el cual ser acutalizada por el descenso por el gradiente
    theta = theta.copy()
    J_history = []
    
    for i in range(num_iters):
        h = sigmoid(X.dot(theta.T))
        theta = theta - (alpha / m) * (h - y).dot(X)
       
        J_history.append(calcularCosto(theta, X, y))
    return theta, J_history
```

```python
# Elegir algun valor para alpha (probar varias alternativas)
alpha = 0.01
num_iters = 10000

# inicializa theta y ejecuta el descenso por el gradiente
theta = np.zeros(7)
theta, J_history = descensoGradiente(theta, X, y, alpha, num_iters)

# Grafica la convergencia del costo
pyplot.plot(np.arange(len(J_history)), J_history, lw=2)
pyplot.xlabel('Numero de iteraciones')
pyplot.ylabel('Costo J')
print("costo min: "+ str(J_history[-1]) )

# Muestra los resultados del descenso por el gradiente
print('theta calculado por el descenso por el gradiente: {:s}'.format(str(theta)))

# verificar si ingresa o no a la universidad
X_array = [1,0.38,0.53,2,3,0,0]
aprueba = sigmoid(np.dot(X_array, theta))   # Se debe cambiar esto

print('probabilidad de que un empleado sea dejo o abandono el area es  (usando el descenso por el gradiente):{:.0f}'.format(aprueba))
print("Si la probabilidad es menor a cero o menor a 0.50 (no dejo) \n Si la probabilidad es mayor a 0.50(si dejo) ")
```

```python
def costFunction(theta, X, y):
    # Inicializar algunos valores utiles
    m = y.size  # numero de ejemplos de entrenamiento

    J = 0
    grad = np.zeros(theta.shape)

    h = sigmoid(X.dot(theta.T))
    
    J = (1 / m) * np.sum(-y.dot(np.log(h)) - (1 - y).dot(np.log(1 - h)))
    grad = (1 / m) * (h - y).dot(X)
    
    return J, grad
```

Se prueba la funcion `costFunction` utilizando dos casos de prueba para $\theta$.

```python
# Inicializacion de parametros de ajuste
initial_theta = np.zeros(n+1)
print(initial_theta)
cost, grad = costFunction(initial_theta, X, y)

print('Costo en theta inicial (zeros): {:.3f}'.format(cost))
print('Costo esperado (aproximado): 0.693\n')
print(grad)
print('Gradiente en theta inicial (zeros):')
print('\t[{:.4f}, {:.4f}, {:.4f}]'.format(*grad))
print('Gradiente esperado (aproximado):\n\t[-0.1000, -12.0092, -11.2628]\n')

```

```python
# Calcula y muestra el costo y el gradiente con valores de theta diferentes a cero

test_theta = np.array([-24,0.38,0.53,2,3,0,0])
#test_theta = np.array([-11.74749157, 0.09927308, 0.09316497])
print(test_theta)
cost, grad = costFunction(test_theta, X, y)

print('Costo en theta prueba: {:.3f}'.format(cost))
print('Costo esperado (aproximado): 0.218\n')

print('Gradiente en theta prueba:')
print('\t[{:.3f}, {:.3f}, {:.3f}]'.format(*grad))
print('Gradiente esperado (aproximado):\n\t[0.043, 2.566, 2.647]')
```

#### 1.2.3 Parmetros de aprendizaje usando `scipy.optimize`

En el codigo anterior se encontr los parmetros ptimos de un modelo de regresin lineal al implementar el descenso de gradiente. Se implemento una funcin de costo y se calcul su gradiente, utilizando el algoritmo del descenso por el gradiente.

En lugar de realizar los pasos del descenso por el gradiente, se utilizar el [mdulo `scipy.optimize`] (https://docs.scipy.org/doc/scipy/reference/optimize.html). SciPy es una biblioteca de computacin numrica para `python`. Proporciona un mdulo de optimizacin para la bsqueda y minimizacin de races. A partir de `scipy 1.0`, la funcin` scipy.optimize.minimize` es el mtodo a utilizar para problemas de optimizacin (tanto restringidos como no restringidos).

For logistic regression, you want to optimize the cost function $J(\theta)$ with parameters $\theta$.
Concretely, you are going to use `optimize.minimize` to find the best parameters $\theta$ for the logistic regression cost function, given a fixed dataset (of X and y values). You will pass to `optimize.minimize` the following inputs:

Para la regresin logstica, se desea optimizar la funcin de costo $J(\theta)$ con los parmetros $\theta$.
Concretamente, se va a utilizar `optimize.minimize` para encontrar los mejores parmetros $\theta$ para la funcin de costo de regresin logstica, dado un dataset fijo (de valores X e y). Se pasara a `optimize.minimize` las siguientes entradas:

- `costFunction`: Una funcin de costo que, cuando se le da el dataset de entrenamiento y un $\theta$ particular, calcula el costo de regresin logstica y el gradiente con respecto a $\theta$ para el dataset(X, y). Es importante tener en cuenta que solo se pasa el nombre de la funcin sin el parntesis. Esto indica que solo proporcionamos una referencia a esta funcin y no evaluamos el resultado de esta funcin.
- `initial_theta`: Los valores iniciales de los parametros que se tratan de optimizar.
- `(X, y)`: Estos son argumentos adicionales a la funcion de costo.
- `jac`: Indicacin si la funcin de costo devuelve el jacobiano (gradiente) junto con el valor de costo. (True)
- `method`: Mtodo / algoritmo de optimizacin a utilizar
- `options`: Opciones adicionales que pueden ser especficas del mtodo de optimizacin especfico. Solo se indica al algoritmo el nmero mximo de iteraciones antes de que termine.

Si se ha completado la `costFunction` correctamente,`optimize.minimize` converger en los parmetros de optimizacin correctos y devolver los valores finales del costo y $\theta$ en un objeto de clase.

Al usar `optimize.minimize`, no se tuvo que escribir ningn bucle ni establecer una tasa de aprendizaje como se hizo para el descenso de gradientes. Todo esto se hace mediante `optimize.minimize`: solo se necesita proporcionar una funcin que calcule el costo y el gradiente.

A continuacin, se tiene el cdigo para llamar a `optimize.minimize` con los argumentos correctos.

```python
# Establecer las opciones para optimize.minimize
options= {'maxiter': 1000}

# revisar la documentacion de scipy's optimize.minimize para mayor descripcion de los parametros
# La funcion devuekve un objeto `OptimizeResult`
# Se utiliza el algoritmo de Newton truncado para la optimizacin.
res = optimize.minimize(costFunction,
                        initial_theta,
                        (X, y),
                        jac=True,
                        method='TNC',
                        options=options)

# la propiedad fun del objeto devuelto por `OptimizeResult`
# contiene el valor del costFunction de un theta optimizado
cost = res.fun

# Theta optimizada esta en la propiedad x
theta = res.x

# Imprimir theta en la pantalla
print('Costo con un valor de theta encontrado por optimize.minimize: {:.3f}'.format(cost))
print('Costo esperado (aproximado): 0.203\n');

print('theta:')
print('\t[{:.3f}, {:.3f}, {:.3f}]'.format(*theta))
print('Theta esperado (aproximado):\n\t[-25.161, 0.206, 0.201]')
```

Una vez que se completa `optimize.minimize`, se usa el valor final de $\theta$ para visualizar el lmite de decisin en los datos de entrenamiento.

Para hacerlo, se implementa la funcin `plotDecisionBoundary` para trazar el lmite de decisin sobre los datos de entrenamiento.

```python
def plotDecisionBoundary(plotData, theta, X, y):
    """
    Grafica los puntos X y Y en una nueva figura con un limite de desicion definido por theta.
    the data points X and y into a new figure with the decision boundary defined by theta.
    Grafica los puntos con * para los ejemplos positivos y con o para los ejemplos negativos.
    
    Parametros:
    ----------
    plotData : func
        A function reference for plotting the X, y data.

    theta : array_like
        Parametros para la regresion logistica. Un vecto de la forma (n+1, ).

    X : array_like
        Data set de entrada. Se supone que X es una de las siguientes:
            1) Matriz Mx3, donde la primera columna es una columna de unos para intercepcin.
            2) Matriz MxN, N> 3, donde la primera columna son todos unos.

    y : array_like
        Vector de datos de etiquetas de la forma (m, ).
    """
    # hacer que theta sera un arreglo numpy
    theta = np.array(theta)

    # Graficar los datos (recordar que la primera columna en X es la intercepcin)
    plotData(X[:, 1:7], y)

    if X.shape[1] <= 7:
        # Solo se requieren 2 puntos para definir una linea, para lo cual se eligen dos puntos finales
        plot_x = np.array([np.min(X[:, 1]) - 2, np.max(X[:, 1]) + 2])

        # Calcular la lnea lmite de decisin
        plot_y = (-1. / theta[2]) * (theta[1] * plot_x + theta[0])

        print(plot_x)
        print(plot_y)
        # Graficar y ajustar los ejes para una mejor visualizacin
        pyplot.plot(plot_x, plot_y)

        # Leyenda, especifica para el ejercicio
        pyplot.legend(['Dejo el puesto', 'No dejo el puesto', 'Limite de decisin'])
        pyplot.xlim([2, 10])
        pyplot.ylim([2, 10])
    else:
        # Rango de la grilla
        u = np.linspace(-1, 1.5, 50)
        v = np.linspace(-1, 1.5, 50)

        z = np.zeros((u.size, v.size))
        # Evalua z = theta*x sobre la grilla
        for i, ui in enumerate(u):
            for j, vj in enumerate(v):
                z[i, j] = np.dot(np.mapFeature(ui, vj), theta)

        z = z.T  # importante transponer z antes de llamar al contorno
        # print(z)

        # Plot z = 0
        pyplot.contour(u, v, z, levels=[0], linewidths=2, colors='g')
        pyplot.contourf(u, v, z, levels=[np.min(z), 0, np.max(z)], cmap='Greens', alpha=0.4)
```

```python
# Graficar limites
plotDecisionBoundary(plotData, theta, X, y)
```

#### 1.2.4 Evaluacin de la regresin logistica

Despus de aprender los parmetros, se puede usar el modelo para predecir si un estudiante en particular ser admitido. Para un estudiante con una puntuacin en el Examen 1 de 45 y una puntuacin en el Examen 2 de 85, debe esperar ver una probabilidad de admisin de 0,776. Otra forma de evaluar la calidad de los parmetros que hemos encontrado es ver qu tan bien predice el modelo aprendido en nuestro conjunto de entrenamiento. 

```python
def predict(theta, X):
    """
    Predecir si la etiqueta es 0 o 1 mediante regresin logstica aprendida.
    Calcula las predicciones para X usando un umbral en 0.5 (es decir, si sigmoide (theta.T * x)> = 0.5, predice 1)

    Parametros
    ----------
    theta : array_like
        Parametros para regresion logistica. Un vecto de la forma (n+1, ).
    
    X : array_like
        Datos utilizados para el calculo de las predicciones. 
        La fila es el numero de los puntos para calcular las predicciones, 
        y las columnas con el numero de caracteristicas.

    Devuelve
    -------
    p : array_like
        Predicciones y 0 o 1 para cada fila en X. 
    """
    m = X.shape[0] # Numero de ejemplo de entrenamiento

    p = np.zeros(m)

    p = np.round(sigmoid(X.dot(theta.T)))
    return p
```

Una vez entrenado el modelo se procede a realizar la prediccion y evaluacin de los resultados de predecir cual es el valor que vota el modelo para todos los datos utilizados en el entrenamiento.

```python
#  Predice la probabilidad de ingreso para un estudiante con nota de 45 en el examen 1 y nota de 85 en el examen 2
prob = sigmoid(np.dot([1, 0.38,0.53,2,3,0,0], theta))
#print('probabilidad de que un empleado sea dejo o abandono el area o puesto es del : {:.3f}%'.format(prob))
print('Valor esperado: -1.91 +/-  3. \n')
print("-Si la probabilidad es menor a cero o menor a 0.50 o 50% (no dejo) \n -Si la probabilidad es mayor a 0.50 o 50%(si dejo) ")

print("\nprobabilidad de que un empleado sea dejo o abandono el area o puesto es del :"+str( prob ))
# Compute accuracy on our training set
p = predict(theta, X)
print('Precisin de entrenamiento: {:.2f} %'.format(np.mean(p == y) * 100))
print('\nPrecisin esperada (aproximadamente): 0.3 ')
```
